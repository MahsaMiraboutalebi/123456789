```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})


df = pd.read_csv('travel_times.csv')

print('Printing the shape of the dataset:')
shape = df.shape
print(shape, "\n")

print('Printing the first 15 rows of the dataset:')
print(df.head(15), "\n")

print('Getting information for the features (columns) with missing values:')
df.info()
print('\n')

print('Dropping duplicate values (if any) by keeping only the last instance:')
df.drop_duplicates(inplace=True, keep='last')
print(shape, "\n")

print('Calculating the total number of missing values (if any) on each column:')
print(df.isnull().sum(), "\n")

print('Dropping rows with missing values from the 1st copy:')
row_df = df.copy()
print(row_df.dropna(), '\n')

print('Dropping columns with missing values from the 2nd copy:')
column_df = df.copy()
print(column_df.dropna(axis=1),'\n')

print('Getting summary statistics:')
print(df.describe(),'\n')

print('Viewing the correlation between the numerical columns:')
print(df.corr,'\n')

print('Showing rows 11 to 14:')
print(df.iloc[11:15],'\n')

print('Creating a subset with trips occurred on November 23, 2011 and January 6, 2012:')
print(df[(df['Date'] == '11/23/2011') | (df['Date'] == '1/6/2012')], '\n')

print('Producing plots for dataset:')
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})

df.plot(kind='scatter', x='TotalTime', y='Distance', title='Total Time vs. Distance')

fig, axes = plt.subplots(nrows=1, ncols=1)
df['AvgSpeed'].plot(kind='box', title='Average Speed')

fig, axes = plt.subplots(nrows=1, ncols=1)
df['AvgMovingSpeed'].plot(kind='box', title='Average Moving Speed')
```
